## 동전 던지기

```python
import random
```

```python
current_cnt = 0
prob_list = []
success_cnt = 0
x_index_list = []
for i in range(1000) :
    x_index_list.append(i)
    result = random.choice(['H', 'T'])
    if result == 'H' :
        success_cnt += 1
    else :
        pass
    current_cnt += 1
    probability = success_cnt / current_cnt
    prob_list.append(probability)
    
```

```python
import matplotlib.pyplot as plt

plt.plot(x_index_list, prob_list, marker="o")
plt.axhline(y=0.5, color='r', linestyle='-')
result_prob = prob_list[-1]
plt.title(f"Probability is {result_prob}")
```
